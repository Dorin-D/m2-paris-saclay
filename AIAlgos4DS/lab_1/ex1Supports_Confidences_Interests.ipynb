```python
import itertools
```

```python
baskets = [set("mcb"), 
           set("mpj"), 
            set("mb"), 
            set("cj"), 
            set("mpb"), 
            set("mcbj"), 
            set("cbj"), 
            set("bj")]

#association rules: mb, bc, cj

permutations = list(itertools.permutations(baskets, 4))
for permutation in permutations:
    print(permutation)
```

```python
#support("mb")
support_mb = sum([1 for basket in baskets if set("mb").issubset(basket)])
support_bc = sum([1 for basket in baskets if set("bc").issubset(basket)])
support_cj = sum([1 for basket in baskets if set("cj").issubset(basket)])

#support m,b,c,j
support_m = sum([1 for basket in baskets if set("m").issubset(basket)])
support_b = sum([1 for basket in baskets if set("b").issubset(basket)])
support_c = sum([1 for basket in baskets if set("c").issubset(basket)])
support_j = sum([1 for basket in baskets if set("j").issubset(basket)])
```

```python
print("Supports: mb = {}, bc = {}, cj = {}".format(support_mb, support_bc, support_cj))
print("Supports: m = {}, b = {}, c = {}, j = {}".format(support_m, support_b, support_c, support_j))
```

```python
#confidence m if b in basket
confidence_mb = support_mb / support_b
#confidence b if m in basket
confidence_bm = support_mb / support_m
#confidence b if c in basket
confidence_bc = support_bc / support_c
#confidence c if b in basket
confidence_cb = support_bc / support_b
#confidence c if j in basket
confidence_cj = support_cj / support_j
#confidence j if c in basket
confidence_jc = support_cj / support_c
print("Confidence m if b: ", confidence_mb)
print("Confidence b if m: ", confidence_bm)
print("Confidence b if c: ", confidence_bc)
print("Confidence c if b: ", confidence_cb)
print("Confidence c if j: ", confidence_cj)
print("Confidence j if c: ", confidence_jc)
```

```python
interest_mb = confidence_mb - support_m/len(baskets)
print("interest mb = ", interest_mb)
interest_bm = confidence_bm - support_b/len(baskets)
print("interest bm = ", interest_bm)
interest_bc = confidence_bc - support_b/len(baskets)
print("interest bc = ", interest_bc)
interest_cb = confidence_cb - support_c/len(baskets)
print("interest cb = ", interest_cb)
interest_cj = confidence_cj - support_c/len(baskets)
print("interest cj = ", interest_cj)
interest_jc = confidence_jc - support_j/len(baskets)
print("interest jc = ", interest_jc)
```
