```python
import numpy as np
from matplotlib import pyplot as plt
import scipy 
```

```python
#read 'barbara.jpg'
img = plt.imread('barbara.jpg')
#convert to grayscale
img = img.mean(axis=2)
```

```python
plt.imshow(img, cmap='gray')
```

Gradient filter

Sobel filter

```python
sobel_vertical_filter = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_horizontal_filter = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

#apply sobel filter to image
image_sobel_vertical = scipy.signal.convolve2d(img, sobel_vertical_filter, mode='same', boundary='fill', fillvalue=0)
image_sobel_horizontal = scipy.signal.convolve2d(img, sobel_horizontal_filter, mode='same', boundary='fill', fillvalue=0)
```

```python
plt.imshow(image_sobel_vertical, cmap='gray')
```

```python
plt.imshow(image_sobel_horizontal, cmap='gray')
```

```python
image_sobel_horizontal.shape
```

Averaging filter

```python
averaging_filter = np.ones((3,3))/9
image_averaging = scipy.signal.convolve2d(img, averaging_filter, mode='same', boundary='fill', fillvalue=0)
plt.imshow(image_averaging, cmap='gray')
```

Gaussian filter

```python
gaussian_image = scipy.ndimage.filters.gaussian_filter(img, sigma=1)
plt.imshow(gaussian_image, cmap='gray')
```
