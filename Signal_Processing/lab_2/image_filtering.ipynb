```python
import numpy as np
from matplotlib import pyplot as plt
import scipy 
```

```python
#read image
img = plt.imread('barbara.jpg')
#convert to grayscale
img = img.mean(axis=2)
```

```python
plt.imshow(img, cmap='gray')
```

Gradient filter

```python
gradient_vertical_filter = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
gradient_horizontal_filter = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

#apply gradient filter to image
image_gradient_vertical = scipy.signal.convolve2d(img, gradient_vertical_filter, mode='same', boundary='fill', fillvalue=0)
image_gradient_output = scipy.signal.convolve2d(image_gradient_vertical, gradient_horizontal_filter, mode='same', boundary='fill', fillvalue=0)
```

```python
plt.imshow(image_gradient_vertical, cmap='gray')
```

```python
plt.imshow(image_gradient_output, cmap='gray')
```

Sobel filter

```python
sobel_vertical_filter = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_horizontal_filter = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

#apply sobel filter to image
image_sobel_vertical = scipy.signal.convolve2d(img, sobel_vertical_filter, mode='same', boundary='fill', fillvalue=0)
image_sobel_output = scipy.signal.convolve2d(image_sobel_vertical, sobel_horizontal_filter, mode='same', boundary='fill', fillvalue=0)
```

```python
plt.imshow(image_sobel_output, cmap='gray')
```

Averaging filter

```python
averaging_filter = np.ones((3,3))/9
image_averaging = scipy.signal.convolve2d(img, averaging_filter, mode='same', boundary='fill', fillvalue=0)
plt.imshow(image_averaging, cmap='gray')
```

```python
#apply the average filter multiple times
repeat = 10
average_image = np.copy(img)
for i in range(repeat):
    average_image = scipy.signal.convolve2d(average_image, averaging_filter, mode='same', boundary='fill', fillvalue=0)

plt.imshow(average_image, cmap='gray')
```

Gaussian filter

```python
gaussian_filter = np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]])/16
#gaussian_image = scipy.ndimage.filters.gaussian_filter(img, sigma=1)
gaussian_image = scipy.signal.convolve2d(img, gaussian_filter, mode='same', boundary='fill', fillvalue=0)
plt.imshow(gaussian_image, cmap='gray')
```

```python
#apply the gaussian filter multiple times
repeat = 10
gaussian_image = np.copy(img)
for i in range(repeat):
    gaussian_image = scipy.signal.convolve2d(gaussian_image, gaussian_filter, mode='same', boundary='fill', fillvalue=0)

plt.imshow(gaussian_image, cmap='gray')
```

Discussion

The gradient filter works similarly to the sobel filter; their function is to detect vertical and horizontal lines.

The average filter and the gaussian filters have the role of blurring images.
