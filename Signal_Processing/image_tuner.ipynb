```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read image content
img = plt.imread('barbara.jpg')
```

```python
#reshape image to grayscale
img = img[:,:,0]
```

```python
#perform fourier transform
img_fft = np.fft.fft2(img)
```

```python
#plot fourier transform
plt.imshow(np.log(np.abs(img_fft)), cmap='gray')
```
